```python
from bs4 import BeautifulSoup as bs
```

```python
search_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8'

res = req.get(search_url,params={'query':'탕수육'})
res
```

```python
# bs(변활할 데이터, 변환도구)
# -웹 문서를 파이썬 자료형으로 변환해주는 기능
# 변환할 데이터 : 웹 문서(HTML)
# 변환도구 : lxml > html.parser > html5lib (변환속도 순)
html = bs(res.text,'lxml')
```

```python
type(res.text), type(html)
```

```python
# CSS선택자
# 1.타입선택자 : 요소 이름
# 2.클래스선택자 : .클래스이름 
# 3.아이디선택자 : #아이디이름
# 4.전체선택자 : * 
# 5.계층선택자 : 자식, 자손, 근접후행, 후행
html.select_one('title')
```

```python
html.select_one('title').text
```

```python
# 데이터 추출 Tip
# 1. 접근한 HTMl요소에 id나 class속성이 있는지 체크
# 2. 없으면, 부모요소의 아이디, class속성 확인 (계층구조 고려)
html.select_one('a.title')
```

```python
html.select_one('a.title').text
```

```python
# 접근한 요소의 속성값 가져오기
# .get(속성이름)
html.select_one('a.title').get('href')
```

```python
# select(css선택자): 선택자 조건에 해당하는 하나의 HTML 요소를 찾아서 반환
# select(css선택자):선택자 조건에 해당하는 모든 HTML요소를 찾아서 리스트로 반환
aTags = html.select('a.title')
```

```python
for i in range(len(aTags)):
    print(aTags[i])
    print(aTags[i].get('href'))
    
```

```python
for i in aTags[2:4] :
    print(i.text)
    print(i.get('href'))
```

#### 뉴스 목록 페이지에 있는 뉴스 제목과 URL주소를 출력
출력 결과
1. 뉴스제목 : ____
2. 바로가기: http://__

```python
newslist_url = 'https://news.naver.com/breakingnews/section/105/731'
res = req.get(newslist_url)
html_nl = bs(res.text,'lxml')
html_nl
```

```python
news_title=html_nl.select('.sa_text_strong')
```

```python
news_title[0].text
```

```python
news_link=html_nl.select('.sa_text_title')
news_link[0].get('href')
```

```python
for i in range(len(news_title)) :
    print(f'뉴스제목 : {news_title[i].text}')
    print('바로가기 :',  end=' ' )
    print(news_link[i].get('href'))
    print('='*80)
 
```

```python
for i in range(len(news_title)) :
    print(f'{i+1}.')
    print(f'뉴스제목 : {news_title[i].text.strip()}')
    print(f"바로가기 : {news_link[i].get('href')}")
    print()
```
