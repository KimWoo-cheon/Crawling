```python
# 웹 브라우저를 통제하는 용도
from selenium import webdriver as wb
# 키보드의 역할
from selenium.webdriver.common.keys import Keys
# 찾을 방법(선택자) 지정하기
from selenium.webdriver.common.by import By
import pandas as pd
import time
```

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')
time.sleep(1)


# 더보기 클릭하기


# more_btn

# more_btn.click()
more_btn = driver.find_element(By.CLASS_NAME,'c_05')
# 예외처리
try:
    while True:
        #페이지가 갱신되면 요소를 다시 찾아야함
        # more_btn = driver.find_element(By.CLASS_NAME,'c_05')
        more_btn.click()
        # time.sleep(1)
except:
    print('더보기 클릭 끝')

time.sleep(3)

menu_list=driver.find_elements(By.CLASS_NAME,'h.fz_03')
menu_list
price_list=driver.find_elements(By.CSS_SELECTOR,'.item-price strong')
price_list
```

```python
menus=[]
prices=[]
for i in range(len(menu_list)) :
    menus.append(menu_list[i].text.strip())
    prices.append(price_list[i].text.strip())

print(len(menus))
print(len(prices))
```

```python
data1 = { "메뉴": menus, "가격" : prices}
df3=pd.DataFrame(data1)
df3
```
