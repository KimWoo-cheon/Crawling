```python
melon_url = 'https://www.melon.com/chart/index.htm'
res=req.get(melon_url, headers=headerInfo)
html=bs(res.text,'lxml')
```

#### 1.멜론 음원 데이터 수집
- 순위, 곡명, 가수
- DataFrame으로 만들기
- 엑셀파일로 저

```python
rank=html.select('.rank')

rank    
    
```

```python
for i in range(len(rank)) :
    rank[i]=rank[i].text

rank
```

```python
title=html.select('.rank01 a')
title[10].text
```

```python
for i in range(len(title)) :
    title[i] = title[i].text
title
```

```python
print(html.select('.rank02')[94].text)
```

```python
singer=html.select('.rank02') #text
singer2=[]
for i in range(0,len(singer),1) :
   singer2.append(singer[i].text[:len(singer[i].text)//2].strip())
singer2
```

```python
len(singer2)
len(title)
len(rank)
```

```python
data = {"순위":rank[1:], "제목" : title, "가수" : singer2}
```

```python
df1=pd.DataFrame(data)
df1
df1.set_index('순위',inplace=True)
df1
```

```python
data2 = {'제목':title, '가수':singer2}
```

```python
df2=pd.DataFrame(data2, index=rank[1:])
df2
```

```python
df2.index.name='순위'
df2
```
